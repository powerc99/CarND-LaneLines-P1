Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=5):
    """
    This function is passed an array of end points for line segments in var lines. The function iterates through the line 
    segments keeping track of the coordinates of the x,y points with the largest and smallest X coordinates for both the 
    right and left side. Line segments on the left half of the image with a slope > -0.4 are discarded. line segments on 
    the right half of the image with a slope < 0.4 are discarded. Once the end points for the right and left side are 
    determined the function uses the slope of each line to calculate a new minimum x coordinates for the left side and a 
    new maximum x coordinate for the right side. A line is then drawn on the right and left side of the image using the 
    calculated points.

    """
    imshape = img.shape # get the dimensions of the image
    max_x_leftside=0 # set max x,y on the left and right  to zero
    max_y_leftside=0
    max_x_rightside=0 
    max_y_rightside=0
    min_x_leftside = imshape[0]  #set the min x, y of the left and right to the largest x,y values of the image
    min_y_leftside= imshape[1]
    min_x_rightside= imshape[1]
    min_y_rightside= imshape[0]
   
    for line in lines:
        for x1,y1,x2,y2 in line:
            if (x2-x1)!=0 and (y2-y1)!=0:  # check for slope of 0 and divide by 0
                if ((y2-y1)/(x2-x1)) < -0.4 and (x1 < imshape[1]/2) and (x2 <imshape[1]/2):
                    if x1> max_x_leftside:      # if the line segement falls on the left half of the image and the slope
                        max_x_leftside = x1     # is less than -0.4 update min and max for x,y on the left
                        max_y_leftside = y1
                    if x2> max_x_leftside:
                        max_x_leftside = x2
                        max_y_leftside = y2
                    if x1< min_x_leftside:
                        min_x_leftside = x1 
                        min_y_leftside = y1
                    if x2< min_x_leftside:
                        min_x_leftside = x2
                        min_y_leftside = y2
                elif ((y2-y1)/(x2-x1)) > 0.4 and (x1 > imshape[1]/2) and (x2 >imshape[1]/2):
                    if x1>max_x_rightside:         # if the line segement falls on the right half of the image and the slope 
                        max_x_rightside = x1       # is greater than 0.4 update min and max for x,y on the right side
                        max_y_rightside = y1
                    if x2> max_x_rightside:
                        max_x_rightside = x2
                        max_y_rightside = y2
                    if x1< min_x_rightside:
                        min_x_rightside = x1 
                        min_y_rightside = y1
                    if x2< min_x_rightside:
                        min_x_rightside = x2
                        min_y_rightside = y2
                        
    # we should now have the endpoints with the largest and smallest x value for each line on the right and left
    
    # calculate the slope of the new line segments on the left and right
    if max_x_rightside- min_x_rightside !=0:
        slope_right = (max_y_rightside-min_y_rightside) / (max_x_rightside- min_x_rightside)
    else:
        slope_right = 0
    if max_x_leftside -min_x_leftside !=0:
        slope_left = (max_y_leftside-min_y_leftside) / (max_x_leftside -min_x_leftside)
    else:
         slope_left = 0
   
    
    if slope_right!=0:   # set y to the botom of the image and calculate a new  max x on the right side using the slope
        max_y_rightside = imshape[0] 
        max_x_rightside = int( ((max_y_rightside - min_y_rightside)/slope_right) + min_x_rightside)
        
    if slope_left!=0:   # set y to the botom of the image and calculate a new  min x on the left side using the slope
        min_y_leftside = imshape[0] 
        min_x_leftside = int((max_y_leftside - min_y_leftside)/ -slope_left) + max_x_leftside
         
    # draw 2 lines from the calculated x,y values above
    cv2.line(img, (min_x_rightside, min_y_rightside ), (max_x_rightside, max_y_rightside), color, thickness)
    cv2.line(img, (min_x_leftside, min_y_leftside ), (max_x_leftside, max_y_leftside), color, thickness)
    
    
def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img



def weighted_img(img, initial_img, =0.8, =1., =0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img *  + img *  + 
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, , img, , )
````

````python
# Import everything needed to edit/save/watch video clips
from moviepy.editor import VideoFileClip
from IPython.display import HTML
````

````python
def process_image(image):
    # NOTE: The output you return should be a color image (3 channel) for processing video below
    # TODO: put your pipeline here,
    # you should return the final output (image with lines are drawn on lanes)
    gray = grayscale(image)

    # Define a kernel size and apply Gaussian smoothing
    kernel_size = 5
    blur_gray = gaussian_blur(gray, kernel_size)

    # Define our parameters for Canny and apply
    low_threshold = 50
    high_threshold = 150
    edges = canny(blur_gray, low_threshold, high_threshold)


    # Next we'll create a masked edges image 
    imshape = image.shape
    
    #define our area of interest
    vertices = np.array([[(0,imshape[0]),(int(imshape[1]*.46), int(imshape[0]*.62)), (int(imshape[1]*.56), int(imshape[0]*.62)), (imshape[1],imshape[0])]], dtype=np.int32)
    masked_edges= region_of_interest(edges, vertices)


    # Define the Hough transform parameters
    rho = 2 # distance resolution in pixels of the Hough grid
    theta = np.pi/180 # angular resolution in radians of the Hough grid
    threshold = 15     # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 40 #minimum number of pixels making up a line
    max_line_gap = 20   # maximum gap in pixels between connectable line segments
   
    # Run Hough on edge detected image
    # Output "lines" is an array containing endpoints of detected line segments
    lines =  hough_lines(masked_edges, rho, theta, threshold, min_line_length, max_line_gap)  

    # Draw the lines on the edge image
    lines_edges=weighted_img(lines,image , =0.8, =1., =0.)

    return lines_edges
````

How could you imagine making your algorithm better / more robust?

The algorithm currently finds the top and bottom coordinates of each line and then calculates the slope based on those coordinates. The algorithm may be more robust if a running average of all "qualified" line segments from previous frames was calculated and used to draw the lines. 


Where will your current algorithm be likely to fail?

The algorithm currently filters out certain line segments based on their slope and position in the image. These filters may be too stringent when testing on other road conditions.  The algorithm also picks up the outer most line on both sides of the image. If the road has double lines then the algorithm would choose the outside lines over the inside lines. Also if there are other marks or lines big enough to be detected by the algorithm this could throw off the calculated end points of the line.   

The algorithm currently fails when there is not enough contrast between the road and the lane marker. In the challenge video, when the road color goes from dark to light the algorithm fails to pick up the lane lines at all

Also if another object like a car was in the defined region in front of the car then the algorithm would most likely fail.
